```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
osudf = []
url = 'https://osu.ppy.sh/rankings/osu/performance'
osuhtml = requests.get(url)
soup = BeautifulSoup(osuhtml.text,'lxml')
items = soup.find_all('tr','ranking-page-table__row')
rank = 0
for item in items:
    rank += 1
    osu_name = item.find('div','ranking-page-table__user-link').select('a')[1].text.strip() #osu玩家名字
    profile_link = item.find('div','ranking-page-table__user-link').select('a')[1].get('href') #osu玩家個人檔案連結
    player_acc = item.select('td')[2].text.strip() #accuracy
    play_count = item.select('td')[3].text.strip() #playcount
    score = item.select('td')[4].text.strip() #score
    SS_count = item.select('td')[5].text.strip() #SS 
    S_count = item.select('td')[6].text.strip()  # S
    A_count = item.select('td')[7].text.strip()   # A
    print('第%d名:'%(rank),osu_name,profile_link,player_acc,play_count,score,SS_count,S_count,A_count)
    
    osulist = pd.DataFrame([{'名次':'第%d名:'%(rank),
                        '玩家名字':osu_name,
                        '玩家連結':profile_link,
                        'accuracy':player_acc,
                        'playcount':play_count,
                        'Score':score,
                        'SS個數':SS_count,
                        'S個數':S_count,
                        'A個數':A_count}])
    osudf.append(osulist)
    
osudf = pd.concat(osudf, ignore_index=True)
osudf.to_excel('osuplayers.xlsx',index=0)
```
